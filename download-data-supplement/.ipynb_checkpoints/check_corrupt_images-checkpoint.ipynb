```python
from PIL import Image
import pandas as pd
from multiprocessing import  Pool
import numpy as np
```

```python
#change to reflect your image directory and annotations file
img_dir = '/domino/datasets/local/snapshot-serengeti-images/snapshotserengeti-unzipped'
annotations_file = '/mnt/dataset_labels.csv'

labels = pd.read_csv(annotations_file, low_memory=False)
labels.head()
```

```python
#check if image loads
def check_image(img):
    try:
        with Image.open(img) as im:
            im.load()
        return False
    except:
        return True
```

```python
#add labels for corrupted images
def check_image_df(df):
    df['corrupt'] = df['image_name'].apply(lambda img: check_image(img_dir+'/'+img))
    return df
```

```python
def parallelize_dataframe(df, func, n_cores=4):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df
```

```python
#save
labels = parallelize_dataframe(labels, check_image_df, n_cores=12)
labels.to_csv('/mnt/dataset_labels_w_corrupt.csv', index=False)
labels.loc[labels['corrupt'] == True].head()
```
